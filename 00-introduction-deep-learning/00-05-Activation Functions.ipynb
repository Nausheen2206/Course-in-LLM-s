# Activation Functions 
## What Are Activation Functions?

When we build a neural network, we are trying to make a computer learn from data.  To do this, the network is made up of small units called neurons.

Each neuron receives some input values, combines them, and produces an output.  But the calculation done inside a neuron is very basic. It is mostly just addition and multiplication.
On its own, this kind of calculation is not enough to solve real problems.

An **activation function** is used to decide what the final output of a neuron should be.  It takes the value produced by the neuron and changes it in a specific way before sending it to the next layer.

In simple words, an activation function answers this question:“Should this neuron pass information forward, and if yes, how much?”.

Without activation functions, a neural network would not be able to learn anything useful and would be restricted to modeling only linear relationships between inputs and outputs.


## Why Are Activation Functions Important?

If we remove activation functions from a neural network, every layer would only perform simple linear calculations.Even if we add many layers, the network would still behave like a single straight-line equation.
This is a problem because real-world data is not linear.

For example:
- House prices do not increase in a straight line as size increases
- Images are made of complex shapes, not simple patterns
- Human language does not follow simple mathematical rules

Activation functions introduce **non-linearity** into the network.  This non-linearity allows the network to learn curves, patterns, and complex relationships.

Because of activation functions, neural networks can:
- Learn from images
- Understand speech and text
- Make accurate predictions from messy, real data


## What Happens Inside a Neuron?

Inside a neuron, two main steps happen:
1. The neuron combines its inputs using weights and a bias  
2. The activation function is applied to this combined value

The activation function controls the final output of the neuron.Different activation functions behave differently, which is why choosing the right one matters.

So activation functions:
- Decide the output of each neuron
- Add non-linearity to neural networks
- Make deep learning possible

Without activation functions, neural networks would be too simple to solve real-world problems.


![Neuron and activation function diagram](https://miro.medium.com/1%2AYQHD47q_To9yy1PTihh4bA.jpeg)


Look at the diagram above.

Think of each arrow on the left as a number going into the neuron.  These numbers come from the data. For example, they could be pixel values from an image or values from a table.Inside the neuron, all these numbers are mixed together into **one single number**.  
This mixing is just a calculation done by the neuron.
Now comes the important part.

The activation function looks at this single number and makes a decision:
- if the number is important, it lets it pass forward
- if the number is not important, it reduces it or blocks it

The number that comes out after this decision is the output of the neuron.This output is then sent to the next neuron, where the same steps happen again.


## Activation Functions and Their Main Focus
Each activation function is designed to solve a specific learning problem inside a neural network.

| Activation Function | What It Mainly Focuses On |
|--------------------|---------------------------|
| Linear | Passing values directly without any change |
| Sigmoid | Making yes/no decisions using probabilities |
| Tanh | Balancing positive and negative values around zero |
| ReLU | Speeding up learning by keeping positive values |
| Leaky ReLU | Avoiding inactive (dead) neurons |
| PReLU | Learning how much negative information to allow |
| ELU | Smoother learning for negative inputs |
| Softplus | Smooth version of ReLU without sharp corners |
| Swish | Allowing smooth and flexible information flow |
| Softmax | Selecting one class out of many using probabilities |


## 1. Linear Activation Function

The linear activation function also known as the identity function, is the most basic form of activation function used in neural networks.In this function, the output is directly proportional to the input.  
Mathematically, it can be written as:

 #### **f(x) = x**

where:
    
#### **x=$\sum input*weight+bias$**


This means that whatever value is given as input to the neuron, the same value is produced as output.

The graph of this function is a straight line passing through the origin.Because of this, it is called a *linear* activation function.
![Linear Activation Function](https://media.geeksforgeeks.org/wp-content/uploads/20241029115212560858/Linear-Activation-Function.png)

### Effect on Neural Networks
A neural network performs a series of weighted sums followed by activation functions.If every layer in a neural network uses only linear activation functions, then the entire network behaves like a single linear equation, even if it has many layers.As a result, the network can only learn linear relationships between input and output.
### Output Range
The output of a linear activation function is unbounded.  
It can take any value from:$-\infty$ to $+\infty$

### Where Is It Used?
Because linear activation does not introduce non-linearity:
- It is *not suitable for hidden layers*
- It is sometimes used in the *output layer* for regression problems

Examples include:
- Predicting prices
- Predicting temperature values
- Predicting any continuous numerical output

### Advantages
-  The linear activation function follows a straightforward mathematical rule, which makes it easy to interpret and analyze.

-  It requires very little computation, which makes it fast to calculate during training and inference.

-  The output can take any value from$-\infty$ to $+\infty$, which is useful for problems where predictions are not restricted to a fixed range.

-  Linear activation is well-suited for output layers in regression tasks where the goal is to predict continuous numerical values.

-  Since the function is linear, gradients do not vanish or explode due to the activation function itself.

### Disadvantages
- Linear activation functions do not introduce non-linearity. As a result, the network can only learn straight-line (linear) relationships.

-  Even if many layers are stacked together, a network using only linear activation functions behaves like a single linear equation.

- Hidden layers are meant to extract complex features. Linear activation does not provide this capability.

-  Most real-world data is non-linear in nature, making linear activation insufficient for most deep learning tasks.
### Example
The below code shows that a linear activation function simply passes the input forward without modification, resulting in a straight-line graph.




```python
import numpy as np
from matplotlib import pyplot as plt

# Linear activation Function
def linear(x):
    ''' y = f(x) It returns the input as it is'''
    return x


x_data = np.linspace(-5,5,100)
y_data = linear(x_data)


plt.plot(x_data, y_data)
plt.title('Linear Activation Function')
plt.grid()
plt.legend(['f(x) = x'])
plt.show()



```

Using tensorflow

```python

import tensorflow as tf
import numpy as np

#  (y = 2x)
X = np.array([[1], [2], [3], [4], [5]], dtype=float)
y = np.array([[2], [4], [6], [8], [10]], dtype=float)

model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),   
    tf.keras.layers.Dense(1)      
])

model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
    loss='mse'
)

model.fit(X, y, epochs=200, verbose=0)

x_test = np.array([[6]], dtype=float)
prediction = model.predict(x_test)

print("Prediction for x = 6:", prediction)

weights, bias = model.layers[0].get_weights()
print("Learned weight:", weights)
print("Learned bias:", bias)
```

The model learns a linear relationship of the form \( y = wx + b \).  
The learned weight is close to 2, indicating the model has captured the pattern \( y \approx 2x \).  
For input \( x = 6 \), the predicted value is close to 12.  
Small differences from the exact value occur due to iterative training and limited epochs.


## Sigmoid Activation Function

The Sigmoid activation function is a widely used activation function in neural networks. It maps any real-valued input into a range between 0 and 1, which makes it suitable for representing probabilities.

Mathematically, the Sigmoid activation function is defined as:

#### **Sigmoid(x) = $\frac{1}{1 + e^{-x}}$**

where:

#### **x = $\sum input \times weight + bias$**

As the input value increases, the output of the Sigmoid function approaches 1. As the input value decreases, the output approaches 0. This smooth transition gives the Sigmoid function its characteristic S-shaped curve.

### Graph of Sigmoid Activation Function

![Sigmoid Activation Function](https://media.geeksforgeeks.org/wp-content/uploads/20190911181329/Screenshot-2019-09-11-18.05.46.png)

The graph shows that the Sigmoid function smoothly compresses large positive and negative values into a small output range.

### Effect on Neural Networks

The Sigmoid function introduces non-linearity, allowing neural networks to learn complex patterns. It is commonly used in binary classification tasks where outputs need to be interpreted as probabilities.

However, when input values are very large or very small, the gradient becomes close to zero. This slows down learning during backpropagation and is known as the vanishing gradient problem.

### Output Range

The output of the Sigmoid activation function is bounded.

- Minimum value: 0  
- Maximum value: 1

### Where Is It Used?

Sigmoid activation is typically used:
- In the output layer of binary classification models
- In logistic regression
- When probability-based predictions are required

It is generally not recommended for hidden layers in deep neural networks.

### Advantages

- Outputs lie between 0 and 1
- Smooth and differentiable everywhere
- Easy to interpret as probabilities
- Simple mathematical formulation

### Disadvantages

- Suffers from vanishing gradient problem
- Outputs are not zero-centered
- Slower convergence in deep networks
- Not suitable for most hidden layers

#### Example
The below code shows that the Sigmoid activation function compresses input values into a range between 0 and 1, producing an S-shaped curve that represents probability-like outputs.


```python
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(-10, 10, 100)
z = 1/(1 + np.exp(-x))

plt.plot(x, z)
plt.xlabel("x")
plt.ylabel("Sigmoid(X)")
plt.grid()

plt.show()
```

Using tensorflow


```python
import tensorflow as tf

# Input values
x = tf.constant([-2.0, -1.0, 0.0, 1.0, 2.0])

# Sigmoid activation
y = tf.nn.sigmoid(x)

print(y)
```

From the output, we can understand how the Sigmoid activation function transforms input values.  
It shows that all inputs are compressed into a range between 0 and 1.  
Negative inputs result in values closer to 0, while positive inputs result in values closer to 1.  
This demonstrates why Sigmoid is useful for producing probability-like outputs in neural networks.


## 3. Tanh Activation Function

The tanh activation function, short for hyperbolic tangent, is a non-linear activation function that is closely related to the sigmoid function.

Like sigmoid, tanh compresses the input values into a limited range. However, instead of producing values between 0 and 1, tanh produces values between −1 and 1. Because of this, tanh is often considered an improved version of sigmoid.

Mathematically, the tanh activation function is defined as:

#### f(x) = (eˣ − e⁻ˣ) / (eˣ + e⁻ˣ)

Here, x represents the total input received by a neuron and is given by:

#### x = Σ(input × weight) + bias

When this value x is passed through the tanh function, large positive values of x are pushed close to +1, while large negative values of x are pushed close to −1. Values of x close to zero produce outputs close to 0.

Because of this behavior, the tanh function centers its output around zero.

The graph of the tanh function has an S-shaped curve, similar to the sigmoid function, but it is symmetric around the origin.

![Tanh Activation Function](https://media.geeksforgeeks.org/wp-content/uploads/20250214171817652462/tanh.png)

### Effect on Neural Networks

The tanh activation function introduces non-linearity into the neural network, allowing it to learn complex patterns in data.

Since the output of tanh is zero-centered, positive and negative values are balanced around zero. This often helps the optimization process converge faster compared to sigmoid, especially in shallow networks.

However, tanh still suffers from the vanishing gradient problem. When the input values become very large or very small, the slope of the tanh curve becomes very small. During backpropagation, this leads to very small gradient values, which slows down learning in deeper networks.

### Output Range

The output of the tanh activation function is bounded.

It always lies between −1 and 1.

### Where Is It Used?

Tanh activation has been widely used in hidden layers of neural networks, especially before ReLU became popular.

Typical applications include:
- Hidden layers of older feedforward neural networks
- Some recurrent neural network architectures

### Advantages

- Tanh produces outputs that are centered around zero, which helps reduce bias in weight updates during training.

- It introduces non-linearity, allowing the network to model complex relationships.

- Compared to sigmoid, tanh often provides better gradient flow in practice.

### Disadvantages

- Tanh suffers from the vanishing gradient problem in deep networks.

- Learning becomes slow when input values fall into the saturation regions of the function.

- Because of these limitations, tanh is less commonly used in modern deep neural networks compared to ReLU.


### Derivative of Tanh Activation Function

To train a neural network, we use a process called backpropagation.  
During this process, the network needs to know **how much the output of a neuron changes when its input changes**. This is given by the derivative of the activation function.

The derivative of the tanh activation function is simple and has a very useful form.

If the activation output is:
#### tanh(x)

then its derivative is:

#### 1 − tanh²(x)

This means that the derivative depends directly on the output of the tanh function itself.

When the output value of tanh(x) is close to 0, the derivative is close to 1.  This allows gradients to flow well during training.

However, when the output of tanh(x) is close to −1 or +1, the derivative becomes very small.  In these regions, the gradient almost disappears.

Because of this, learning slows down when neurons enter these saturation regions.

This behavior explains why tanh still suffers from the vanishing gradient problem, especially in deep neural networks.


## Example
This plot shows how the tanh activation function saturates at extreme values while its derivative becomes small, explaining why learning slows in deep networks.



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 200)

tanh_x = np.tanh(x)
tanh_derivative = 1 - tanh_x**2

plt.plot(x, tanh_x, label="tanh(x)")
plt.plot(x, tanh_derivative, label="Derivative of tanh(x)")

plt.xlabel("x")
plt.ylabel("Value")
plt.title("Tanh Activation Function and Its Derivative")
plt.legend()
plt.grid()

plt.show()
```

Using Tensorflow

```python
import tensorflow as tf

# Input values
x = tf.constant([-2.0, -1.0, 0.0, 1.0, 2.0])

# Tanh activation
y = tf.nn.tanh(x)

print(y)
```

From the output, we can understand how the Tanh activation function transforms input values.  
It maps all inputs into a range between −1 and 1.  
Negative inputs produce outputs close to −1, while positive inputs produce outputs close to 1.  
This shows that Tanh is zero-centered and useful for learning balanced representations in neural networks.


## 4. ReLU Activation Function

The ReLU activation function, short for Rectified Linear Unit, is one of the most widely used activation functions in modern deep learning.

Unlike sigmoid and tanh, ReLU does not squash values into a small range. Instead, it removes all negative values and keeps positive values unchanged.

Mathematically, the ReLU function is defined as:

#### f(x) = max(0, x)

Here, x represents the total input received by a neuron and is given by:

#### x = Σ(input × weight) + bias

When this value x is passed through the ReLU function, any negative value becomes 0, while any positive value remains the same.

This simple behavior makes ReLU easy to compute and very effective in practice.

The graph of the ReLU function is a straight line for positive values and zero for negative values.

![ReLU Activation Function](https://media.geeksforgeeks.org/wp-content/uploads/20250129162127770664/Relu-activation-function.png)

Because ReLU does not saturate for positive inputs, it allows gradients to flow more easily during training.

### Effect on Neural Networks

ReLU introduces non-linearity while avoiding the vanishing gradient problem seen in sigmoid and tanh for positive values.

This allows deep neural networks to train faster and more efficiently.

However, for negative input values, the gradient becomes zero. If a neuron consistently receives negative inputs, it may stop learning entirely. This issue is known as the dying ReLU problem.

### Output Range

The output of the ReLU activation function ranges from:

0 to +∞

### Where Is It Used?

ReLU is commonly used in the hidden layers of deep neural networks.

Typical applications include:
- Convolutional neural networks
- Deep feedforward networks
- Most modern deep learning architectures

### Advantages

- ReLU is computationally simple and fast to evaluate.

- It reduces the vanishing gradient problem for positive input values.

- It allows neural networks to converge faster during training.

### Disadvantages

- ReLU outputs zero for all negative inputs.

- Neurons can become inactive if they stop producing positive outputs.

- This can reduce the learning capacity of the network in some cases.


## Example 
The given example below shows how ReLU functionn blocks negative values

```python

import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

x = np.linspace(-6, 6, 100)
y = relu(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.title("ReLU Activation Function")
plt.grid()
plt.show()

```

Using tensorflow

```python
import tensorflow as tf


x = tf.constant([-2.0, -1.0, 0.0, 1.0, 2.0])


y = tf.nn.relu(x)

print(y)
```

From the output, we can understand how the ReLU activation function works.  
All negative input values are converted to 0.  
Positive input values remain unchanged.  
This shows that ReLU introduces non-linearity while keeping computations simple.


## 5. Leaky ReLU Activation Function

The Leaky ReLU activation function is a small improvement over the standard ReLU activation function.

In ReLU, all negative input values are converted to zero. While this helps learning in many cases, it can also cause a problem where some neurons stop learning completely. This issue is known as the dying ReLU problem.

Leaky ReLU was introduced to reduce this problem.

Instead of setting all negative values to zero, Leaky ReLU allows a small negative value to pass through.

Mathematically, the Leaky ReLU function is defined as:

##### f(x) = x, if x > 0  
##### f(x) = αx, if x ≤ 0

Here, α (alpha) is a small positive constant, usually set to a value like 0.01.

The input x represents the total input received by a neuron and is given by:

x = Σ(input × weight) + bias

When this value x is passed through the Leaky ReLU function, positive values remain unchanged, while negative values are scaled down instead of being completely removed.

The graph of the Leaky ReLU function looks similar to ReLU, but the negative side has a small slope instead of being flat.

![Leaky ReLU Activation Function](https://media.geeksforgeeks.org/wp-content/uploads/20200516163110/lerelu.jpg)

### Effect on Neural Networks

Leaky ReLU introduces non-linearity while ensuring that neurons receiving negative inputs can still learn.
Because negative inputs are not completely blocked, gradients can still flow during backpropagation. This reduces the risk of neurons becoming inactive.
As a result, Leaky ReLU often performs better than standard ReLU in situations where dying neurons are a concern.

### Output Range

The output of the Leaky ReLU activation function ranges from:

-$\infty$ to $\infty$

Negative values are scaled by a small factor, while positive values pass through unchanged.

### Where Is It Used?

Leaky ReLU is used in hidden layers of neural networks as an alternative to ReLU.

It is especially useful when:
- ReLU neurons frequently become inactive
- The network struggles to learn due to zero gradients

### Advantages

Leaky ReLU reduces the dying ReLU problem.

It allows gradients to flow even for negative input values.

It is computationally simple and easy to implement.

### Disadvantages

The value of α must be chosen manually.

If α is too small, the function behaves very similar to ReLU.

If α is too large, negative values may affect learning too much.


### Difference Between ReLU and Leaky ReLU

| Feature | ReLU | Leaky ReLU |
|-------|------|------------|
| Output for x > 0 | x | x |
| Output for x < 0 | 0 | αx (small negative value) |
| Slope for x < 0 | 0 | Small constant (e.g., 0.01) |
| Risk of dead neurons | High | Low |
| Learning capability | Stops learning for negative inputs | Continues learning for all inputs |
| Gradient for negative values | Zero | Small but non-zero |
| Effect on backpropagation | Gradients may vanish for negative inputs | Gradients can still flow |
| Use of parameter | No extra parameter | Uses α (leak factor) |
| Computational complexity | Very simple | Slightly more complex |
| Practical usage | Common but may cause dead neurons | Preferred when dead ReLU is an issue |


## Example
This plot shows that Leaky ReLU allows a small negative output for x < 0 instead of blocking values completely.



```python
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu(x, alpha=0.02):
    return np.where(x > 0, x, alpha * x)

x = np.linspace(-6, 6, 100)
y = leaky_relu(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("Leaky ReLU(x)")
plt.title("Leaky ReLU Activation Function")
plt.grid()
plt.show()
```

Using Tensorflow

```python
import tensorflow as tf

x = tf.constant([-2.0, -1.0, 0.0, 1.0, 2.0])

y = tf.nn.leaky_relu(x, alpha=0.1)

print(y)
```

From the output, we can understand how the Leaky ReLU activation function works.  
Positive input values remain unchanged.  
Negative input values are scaled by a small constant instead of being set to zero.  
This helps reduce the dead neuron problem seen in standard ReLU.


## 6. PReLU Activation Function

The PReLU activation function, short for Parametric Rectified Linear Unit, is a learnable version of the ReLU family of activation functions.

To understand PReLU, it helps to recall the limitation of ReLU and Leaky ReLU.  
In standard ReLU, all negative input values are converted to zero, which can cause some neurons to stop learning completely. Leaky ReLU improves this by allowing a small negative slope, but the value of this slope (α) is fixed and chosen manually.

PReLU removes this limitation by allowing the neural network to **learn the value of α automatically during training**.

Mathematically, the PReLU function is defined as:

f(x) = x, if x > 0  
f(x) = αx, if x ≤ 0  

Here, α is not a fixed constant. It is a **trainable parameter**, just like weights and bias, and its value is updated using backpropagation.

The input x represents the weighted sum of inputs to a neuron and is given by:

x = Σ(input × weight) + bias

When x is positive, the PReLU function behaves exactly like ReLU.  
When x is negative, the output is scaled by α, and the amount of scaling is learned from the data.

Because α is learned, different neurons can adapt differently depending on the patterns present in the data.

### Effect on Neural Networks

PReLU introduces non-linearity while ensuring that neurons receiving negative inputs continue to learn.

Unlike ReLU, where the gradient for negative inputs is zero, PReLU maintains a non-zero gradient for negative values. This allows gradients to flow backward during training and significantly reduces the risk of dead neurons.

Since the value of α is learned, the network can decide:
- whether negative values should be strongly suppressed
- weakly suppressed
- or allowed to pass almost unchanged

This adaptability often leads to better performance, especially in deep neural networks.

### Output Range

The output of the PReLU activation function ranges from:

−∞ to +∞

Positive input values pass through unchanged, while negative input values are scaled by the learned parameter α.

### Where Is It Used?

PReLU is mainly used in the hidden layers of deep neural networks.

It is particularly useful when:
- ReLU causes a large number of inactive neurons
- A fixed leak value in Leaky ReLU is not optimal
- The model is deep and requires flexible gradient flow

PReLU has been successfully used in deep convolutional neural networks for image-related tasks.

### Derivative of PReLU

During backpropagation, the derivative of the PReLU function is required.

The derivative is defined as:

f′(x) = 1, if x > 0  
f′(x) = α, if x ≤ 0  

Since α is learnable, its gradient is also computed and updated during training. This allows the network to adjust both weights and activation behavior simultaneously.

### Advantages

PReLU reduces the dying ReLU problem more effectively than standard ReLU.

It allows continuous learning for negative input values.

The learnable parameter α provides flexibility and adaptability.

It often improves convergence and performance in deep networks.

### Disadvantages

PReLU introduces additional trainable parameters into the model.

This slightly increases computational cost.

There is a higher risk of overfitting, especially when training data is limited.

The model becomes more complex compared to ReLU and Leaky ReLU.

### Key Insight

PReLU does not assume how negative values should be treated.  
Instead, it allows the neural network to **learn this behavior directly from data**.


## Example
This plot shows that PReLU behaves like ReLU for positive inputs while allowing a learnable slope for negative inputs.


```python
import numpy as np
import matplotlib.pyplot as plt

def prelu(x, alpha):
    return np.where(x > 0, x, alpha * x)

x = np.linspace(-6, 6, 100)
alpha = 0.2

y = prelu(x, alpha)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("PReLU(x)")
plt.title("PReLU Activation Function")
plt.grid()
plt.show()
```

Using Tensorflow

```python
import tensorflow as tf


x = tf.constant([-2.0, -1.0, 0.0, 1.0, 2.0])


prelu = tf.keras.layers.PReLU()


y = prelu(x)

print(y)
```

From the output, we can understand how the PReLU activation function works.  
Positive input values remain unchanged, similar to ReLU.  
Negative input values are scaled by a learnable parameter instead of being set to zero.  
This allows the network to learn the best negative slope and reduces the dead neuron problem.


## 7. ELU Activation Function

The ELU activation function, short for Exponential Linear Unit, is another improvement over the ReLU family of activation functions.

ReLU and its variants work well for positive input values, but they either block or linearly scale negative values. ELU takes a different approach for negative inputs by using an exponential function instead of a straight line.

The goal of ELU is to improve learning stability and speed by producing smoother outputs for negative input values.

Mathematically, the ELU function is defined as:

#### f(x) = x, if x > 0  
#### f(x) = α(eˣ − 1), if x ≤ 0  

Here, α (alpha) is a positive constant that controls the value to which ELU saturates for negative inputs.

The input x represents the total input received by a neuron and is given by:

x = Σ(input × weight) + bias

When x is positive, ELU behaves exactly like ReLU.  
When x is negative, the output decreases smoothly and approaches −α instead of dropping sharply to zero.

Because of this smooth transition, ELU avoids sudden changes in output values.

The graph of the ELU function shows a straight line for positive inputs and a smooth exponential curve for negative inputs.

![ELU Activation Function](https://media.geeksforgeeks.org/wp-content/uploads/20250603154601294973/Elu_Activation_Function.png)
### Characteristics of the ELU Graph

#### For x > 0

When the input value x is greater than 0, the ELU activation function behaves like the identity function.

In this region:
- The output increases linearly with the input
- The slope of the curve is 1
- The function follows the equation f(x) = x

This part of the ELU graph appears as a straight diagonal line and looks very similar to the ReLU function for positive values.


#### For x ≤ 0

When the input value x is less than or equal to 0, the ELU function behaves differently.

In this region:
- The output decreases smoothly in the negative direction
- The function follows the equation f(x) = α(eˣ − 1)
- The curve does not become flat or suddenly drop to zero

Instead of cutting off negative values like ReLU, ELU gradually approaches a minimum value of −α.  
When α = 1, this minimum value is −1.

The transition at x = 0 is smooth, continuous, and differentiable, which helps maintain stable gradient flow during training.


### Effect on Neural Networks

ELU introduces non-linearity while keeping the mean activation values closer to zero.

Because negative outputs are allowed, ELU helps reduce bias shifts during training, which can improve convergence speed.

The smooth curve for negative values also helps gradients flow more steadily during backpropagation compared to ReLU.

However, ELU still saturates for very negative input values, which can slow learning in those regions.

### Output Range

The output of the ELU activation function ranges from:

$-\alpha$ to $+\infty$.

Positive values pass through unchanged, while negative values approach −α smoothly.

### Where Is It Used?

ELU is mainly used in hidden layers of deep neural networks.

It is especially useful when:
- Faster convergence is required
- Zero-centered activations are preferred
- ReLU-based models show unstable learning

ELU has been used in deep convolutional neural networks.

### Advantages

- ELU produces smooth outputs for negative input values.

- It allows negative outputs, keeping activations closer to zero on average.

- It often leads to faster and more stable training compared to ReLU.

- It reduces the dying ReLU problem.

### Disadvantages

- ELU is more computationally expensive than ReLU and Leaky ReLU.

- It still suffers from saturation for very negative inputs.

- The choice of α can affect performance and usually needs tuning.


## Example
This plot shows that ELU behaves linearly for positive inputs and smoothly approaches a negative saturation value for negative inputs.


```python
import numpy as np
import matplotlib.pyplot as plt

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

x = np.linspace(-6, 6, 200)
y = elu(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("ELU(x)")
plt.title("ELU Activation Function")
plt.grid()
plt.show()
```

## 8. Softmax Activation Function

The Softmax activation function is used mainly in the output layer of neural networks for multi-class classification problems.

In many real-world tasks, a model needs to choose one class out of several possible classes. For example, an image may belong to one of many categories, or a word may belong to one of several parts of speech. In such cases, the model must not only predict a class but also indicate how confident it is about each possible class.

Softmax helps achieve this by converting raw output values, called logits, into probabilities.

Unlike activation functions such as ReLU or tanh that work independently on each neuron, Softmax considers all output neurons together and produces a probability distribution across them.

### Mathematical Definition

The Softmax function is defined as:

#### Softmax(zᵢ) = $\frac{e^{z_i}}{\sum_{j} e^{z_j}}$


Here:
- zᵢ is the raw score (logit) for class i
- The denominator is the sum of exponentials of all class scores

Each output value depends on all the input values, which ensures that the probabilities are normalized.

### How Softmax Works Step by Step

First, the neural network produces a set of raw scores for each class. These scores can be positive, negative, or zero, and they are not probabilities.

Next, the exponential function is applied to each score. This has two effects:
- It makes all values positive
- It magnifies differences between larger and smaller values

Then, each exponential value is divided by the sum of all exponential values. This step normalizes the outputs so that their total becomes exactly 1.

As a result, the final outputs represent probabilities for each class.

A higher input score leads to a higher probability, while lower scores result in smaller probabilities.

### Softmax Output Interpretation

After applying Softmax:
- Each output value lies between 0 and 1
- All output values together sum to 1
- The class with the highest probability is chosen as the model’s prediction

This makes Softmax especially suitable for problems where only one class is correct.

Softmax Activation Function

### Role of Softmax in Neural Networks

Softmax plays a crucial role in classification models by transforming raw model outputs into meaningful probabilities.

During inference, it allows the model to clearly indicate which class it believes is most likely.

During training, Softmax is commonly combined with the cross-entropy loss function. Together, they provide strong gradients that help the model learn faster and more accurately.

This combination simplifies the gradient calculations and improves numerical stability during backpropagation.

### Effect on Learning and Backpropagation

Softmax is differentiable, which allows gradients to be computed during backpropagation.

However, because each Softmax output depends on all input logits, changing one logit affects the probabilities of all classes. This interdependence helps the model learn relative differences between classes rather than treating each class independently.

To improve numerical stability, Softmax is usually implemented by subtracting the maximum logit value from all logits before applying the exponential function. This prevents very large exponent values and avoids overflow issues.

### Output Range

The output of the Softmax activation function lies between 0 and 1 for each class.

All outputs together sum to exactly 1.

### Where Is It Used?

Softmax is used in the output layer of multi-class classification models.

Typical applications include:
- Image classification
- Digit recognition
- Text classification
- Speech recognition

Softmax is used only when the classes are mutually exclusive.

### Advantages

- Softmax produces outputs that can be directly interpreted as probabilities.

- It highlights the most likely class clearly.

- It works well with cross-entropy loss for efficient training.

- It enables comparison between different class predictions.

### Disadvantages

- Softmax is sensitive to very large input values if not implemented carefully.

- It cannot be used in hidden layers.

- It is not suitable for problems where multiple classes can be correct at the same time.

### Key Insight

Softmax does not simply select the largest value.  
It converts all outputs into a probability distribution, allowing the model to express confidence across all classes.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data
y = iris.target

y_encoded = to_categorical(y)

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)

model = Sequential([
    Dense(8, input_shape=(4,), activation='relu'),
    Dense(3, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=8,
    validation_split=0.2,
    verbose=1
)

sample = np.array([[5.1, 3.5, 1.4, 0.2]])

prediction = model.predict(sample, verbose=0)[0]
predicted_class = np.argmax(prediction)

print("\nPredicted Probabilities (Softmax Output):", prediction)
print("Predicted Class:", iris.target_names[predicted_class])


classes = iris.target_names

plt.bar(classes, prediction)
plt.xlabel("Classes")
plt.ylabel("Probability")
plt.title("Softmax Output Probabilities")
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```

# 9.Softplus Activation Function

##  Introduction to Softplus Activation Function

The Softplus activation function is a smooth and continuous activation function commonly used in neural networks. It is designed as a smooth approximation of the ReLU (Rectified Linear Unit) activation function. While ReLU introduces a sharp transition at zero, Softplus replaces this abrupt change with a gradual curve.

Softplus ensures that all output values are positive and that the function remains differentiable everywhere. This property makes it useful in scenarios where smooth gradient flow is important during training.

##  Mathematical Definition of Softplus

The Softplus activation function is defined as:

#### Softplus(x) = $\ln(1 + e^{x})$


Where:
- \(x\) is the input value
- \(e\) is Euler’s number

This formulation allows Softplus to behave similarly to ReLU for large positive values while remaining smooth for negative inputs.



##  Graphical Representation of Softplus


![Softplus Activation Function](https://media.geeksforgeeks.org/wp-content/uploads/20250214190211227942/SoftPlus-Activation-Function.png)



From the graph, we can observe:
- For large positive values of \(x\), Softplus closely follows the ReLU line.
- For negative values of \(x\), Softplus smoothly approaches zero instead of abruptly becoming zero.
- There is no sharp corner at \(x = 0\), unlike ReLU.

This smooth transition helps maintain stable gradients during backpropagation.

##  Derivative of Softplus

The derivative of the Softplus function is:

#### $\frac{d}{dx}\ln(1 + e^{x}) = \frac{e^{x}}{1 + e^{x}}$

This expression is mathematically identical to the **Sigmoid function**.

### Importance of the Derivative
- The gradient is always non-zero.
- Neurons continue to receive gradient updates even for negative inputs.
- This avoids the dead neuron problem commonly seen in ReLU.



### Comparison with ReLU Activation Function

| Aspect | Softplus | ReLU |
|------|---------|------|
| Differentiability | Differentiable everywhere | Not differentiable at zero |
| Output for negative inputs | Small positive values | Zero |
| Gradient behavior | Smooth, never zero | Zero for negative inputs |
| Risk of dead neurons | Very low | High |
| Computational cost | Higher | Very low |
| Sparsity | No sparse outputs | Produces sparse activations |

Softplus trades computational efficiency for smoothness and stability.


##  Advantages 

- Provides smooth gradients across all input values
- Prevents neurons from becoming inactive
- Improves numerical stability in gradient-based optimization
- Suitable when outputs must remain strictly positive
- Acts as a soft version of ReLU with better mathematical properties

## Disadvantages

- Computationally more expensive due to logarithmic and exponential operations
- Does not produce exact zero outputs, reducing sparsity
- May converge more slowly in very deep networks
- Less commonly used compared to ReLU due to performance considerations

##  Practical Use Cases

Softplus is particularly useful in:
- Probabilistic neural networks
- Models requiring positive outputs
- Situations where smooth optimization is critical
- Avoiding dead neurons in sensitive architectures

In practice, Softplus is often chosen for specialized applications rather than as a default activation function.



This code applies PyTorch’s Softplus activation function to a tensor of input values and prints the resulting transformed outputs

```python
import torch
import torch.nn.functional as F

x = torch.tensor([-2.0, -1.0, 0.0, 1.0, 2.0])

y = F.softplus(x)

print(y)
```

From the output, we can understand how the Softplus activation function behaves.  
All input values are transformed into positive outputs.  
Negative inputs produce small positive values instead of zero.  
This shows that Softplus is a smooth approximation of the ReLU function.


## 10. Swish Activation Function

The Swish activation function is a smooth, non-linear activation function used in neural networks. It was introduced as an alternative to ReLU to improve learning performance, especially in deep neural networks. Unlike ReLU, Swish allows small negative values to pass through instead of setting them strictly to zero.

Mathematically, the Swish activation function is defined as:

#### **Swish(x) = x · σ(x)**

where:

#### **σ(x) = $\frac{1}{1 + e^{-x}}$**

Here, σ(x) represents the Sigmoid function.

This means that the input value is multiplied by its corresponding Sigmoid output, resulting in a smooth and adaptive activation.

The Swish function is continuous and differentiable for all input values, which helps maintain stable gradient flow during training.

### Behavior of the Swish Function

- For large positive values of x, the Sigmoid function approaches 1, so Swish(x) behaves almost like a linear function.
- For values of x close to zero, the function transitions smoothly.
- For negative values of x, Swish produces small negative outputs instead of zero.

This behavior helps prevent neurons from becoming inactive.

### Graph of Swish Activation Function
![Swish Activation Function](https://miro.medium.com/v2/resize:fit:786/format:webp/1*SZ19lV6FT24KtdZsvCOCaA.png)



### Effect on Neural Networks

Swish introduces non-linearity while preserving smooth gradients. This allows neural networks to learn complex patterns more effectively compared to linear activation functions.

Because the gradient is non-zero even for negative inputs, Swish reduces the problem of dead neurons and improves optimization during backpropagation.

### Output Range

The output of the Swish activation function is unbounded on the positive side and bounded on the negative side.

- Negative outputs are small and finite
- Positive outputs range from 0 to $+\infty$

### Where Is It Used?

Swish is commonly used in:
- Hidden layers of deep neural networks
- Convolutional neural networks
- Architectures where smooth optimization is required

### Advantages

- Swish is smooth and differentiable everywhere.
- It maintains non-zero gradients for negative inputs.
- It often performs better than ReLU in deep networks.
- Helps improve convergence during training.
- Allows richer feature representation.

### Disadvantages

- Swish is computationally more expensive than ReLU due to the Sigmoid operation.
- It does not produce sparse outputs.
- Slightly slower during training and inference.
- May not be necessary for simple or shallow networks.
